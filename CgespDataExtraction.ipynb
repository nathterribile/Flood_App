```python
import pandas as pd
import pickle
from datetime import datetime
dia_processamento = '2021_12_02'
```

```python
santana_floods = {}


for i in Df_insisdg['IDX']:
    if i<10000:
    
        try:
            cgesp = pd.read_html(Df_insisdg['URL'].loc[i])
            for k in cgesp:
                floods_by_zone = k[0][0]
                if 'Santana' in floods_by_zone:  
                    santana_floods.update({i:k[0][1]})
                 
        except ValueError:
            continue
```

```python
with open(f'dict_santana_{dia_processamento}.pickle', 'wb') as fp:
    pickle.dump(santana_floods, fp, protocol=pickle.HIGHEST_PROTOCOL)
with open(f'dict_santana_{dia_processamento}.pickle', 'rb') as fp:
    santana_floods_2 = pickle.load(fp)    
```
